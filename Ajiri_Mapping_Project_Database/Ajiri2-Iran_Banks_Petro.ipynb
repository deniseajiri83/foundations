```python
import pandas as pd
df_companies = pd.read_csv("SEO_Companies.csv")
df_companies.head()
```

```python
df_petro_all = df_companies[df_companies['name'].str.contains("پتروشیمی", na=False)]
```

```python
df_petro = df_petro_all[~df_petro_all['name'].str.contains("^ح")]

df_petro['link'].value_counts()
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import time
```

```python
list_of_shareholders = []
def all_banks(row):
    driver = webdriver.Chrome()
    driver.get(row['link'])
    shareholders_input = driver.find_element_by_xpath('//*[@id="shareHolder"]/span[1]')
    shareholders_input.click()
    time.sleep(3)
    from bs4 import BeautifulSoup
    doc=BeautifulSoup(driver.page_source, 'html.parser')
    rows = doc.find_all('tr')
    for row in rows[1:]:
        current = {}
        current["time"] = doc.find('span', attrs = {'id': 'liTime'}).text
        current["symbol"] = doc.find('span', attrs = {'id': 'symbol'}).text
        current["name"] = doc.find('span', attrs = {'id': 'com-title'}).text
        current["shareholders"] = row.find_all('td')[0].text
        current["numb_of_shares"] = row.find_all('td')[1].text
        current["percentage"] = row.find_all('td')[2].text
        current["change"] = row.find_all('td')[3].text
        list_of_shareholders.append(current)
df_petro.apply(all_banks, axis=1)
```

```python
list_of_board_members=[]
def all_banks(row):
    driver = webdriver.Chrome()
    driver.get(row['link'])
    search_input = driver.find_element_by_xpath('//*[@id="boardMember"]/span[1]')
    search_input.click()
    time.sleep(3)
    doc=BeautifulSoup(driver.page_source, 'html.parser')
    rows = doc.find_all('tr')
    for row in rows[1:]:
        current = {}
        current["symbol"] = doc.find('span', attrs = {'id': 'symbol'}).text
        current["name"] = doc.find('span', attrs = {'id': 'com-title'}).text
        current['shareholder'] = row.find_all('td')[0].text
        current['Reg_Number'] = row.find_all('td')[1].text
        current['Board_Member'] = row.find_all('td')[2].text
        current['Former_BM'] = row.find_all('td')[3].text
        current['Title'] = row.find_all('td')[4].text
        current['Position'] = row.find_all('td')[5].text
        current['Degree'] = row.find_all('td')[6].text
        list_of_board_members.append(current)
df_petro.apply(all_banks, axis=1)
```

```python
list_of_contact = []
def all_banks(row):
    driver = webdriver.Chrome()
    driver.get(row['link'])
    address_input = driver.find_element_by_xpath('//*[@id="contactInfo"]/span[1]')
    address_input.click()
    time.sleep(3)
    doc=BeautifulSoup(driver.page_source, 'html.parser')
    rows =doc.find_all("p", class_= "info ng-scope")
    for row in rows:
        current = {}
        current["Contact Info"] = row.text
        current['name'] = doc.find('span', attrs = {'id': 'com-title'}).text
        list_of_contact.append(current)
df_petro.apply(all_banks, axis=1)
```

```python
import pandas as pd
df_addresses = pd.DataFrame(list_of_contact)
df_shareholders = pd.DataFrame(list_of_shareholders)
df_board_members = pd.DataFrame(list_of_board_members)
```

```python
df_board_members['Board_member/Title'] = df_board_members['shareholder'] +"-- "+ df_board_members['Title']
df_shareholders['Shareholders/Percentage'] = df_shareholders['shareholders'] +", "+ df_shareholders['percentage']
```

```python
df_output1 = df_shareholders.groupby('name')['Shareholders/Percentage'].apply(lambda x: "<p>%s</p>" % '</p><p> '.join(x)).reset_index(name='Shareholders_info')
df_output1
df_output1.iloc[0]['Shareholders_info']
```

```python
df_output2 = df_board_members.groupby('name')['Board_member/Title'].apply(lambda x: "<p>%s</p>" % '</p><p> '.join(x)).reset_index(name='Board_Members_info')
df_output2
df_output2.iloc[0]['Board_Members_info']
```

```python
df_output3 = df_addresses.groupby('name')['Contact Info'].apply(lambda x: "<p>%s</p>" % '</p><p> '.join(x)).reset_index(name='Contact_info')
df_output3
df_output3.iloc[0]['Contact_info']
```

```python
combined_df1 = df_output2.merge(df_output1, left_on = 'name', right_on = 'name')
```

```python
combined_df2 = combined_df1.merge(df_output3, left_on = 'name', right_on = 'name')
```
