```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("http://tse.ir/listing.html?section=alphabet&cat=")
```

```python
# search_input = driver.find_element('ng-binding')
# search_input.click()
# search_input.click()
```

```python
from bs4 import BeautifulSoup
doc=BeautifulSoup(driver.page_source, 'html.parser')
```

```python
rows = doc.find_all('tr', class_='ng-scope')
len(rows)
#len is 639
```

```python
rows[0].find('a')['href']
#print(s.decode('unicode_escape').encode('ascii','ignore'))
#how to use regex to get rid of this??
```

```python
list_of_companies = []
for element in rows:
    current = {}
    current['name'] = element.find_all('td', class_='ng-binding')[0].text
    current['symbol'] = element.find_all('a', class_='ng-binding')[0].text
    link = element.find('a')['href']
    #link = element.find_all('a')[0]['href']
    current['link'] = ('http://www.tse.ir'+link)
    list_of_companies.append(current)
```
